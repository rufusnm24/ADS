```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA

# Assuming merged_df is your DataFrame
# Convert 'Incident Date' to datetime
merged_df['Incident Date'] = pd.to_datetime(merged_df['Incident Date'])

# Add 'Year' and 'Month' columns for easier aggregation
merged_df['Year'] = merged_df['Incident Date'].dt.year
merged_df['Month'] = merged_df['Incident Date'].dt.month
merged_df['DayOfWeek'] = merged_df['Incident Date'].dt.day_name()

# 1. Incident Frequency Over Time
yearly_incidents = merged_df.groupby('Year').size()
plt.figure(figsize=(10,6))
yearly_incidents.plot(kind='bar', title='Mass Shootings by Year', color='skyblue')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.show()

# 2. Seasonality and Time of Year (Heatmap: Month vs Year)
monthly_incidents = merged_df.groupby(['Year', 'Month']).size().unstack(fill_value=0)
plt.figure(figsize=(12,8))
sns.heatmap(monthly_incidents, cmap='YlGnBu', cbar_kws={'label': 'Number of Incidents'})
plt.title('Heatmap of Mass Shootings by Month and Year')
plt.xlabel('Month')
plt.ylabel('Year')
plt.show()

# 3. Day of the Week Analysis
day_of_week_incidents = merged_df.groupby('DayOfWeek').size()
day_of_week_incidents = day_of_week_incidents[['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']]
plt.figure(figsize=(10,6))
day_of_week_incidents.plot(kind='bar', title='Mass Shootings by Day of the Week', color='lightcoral')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Incidents')
plt.show()

# 4. Time vs. Victims Killed/Injured
victims_by_year = merged_df.groupby('Year').agg({'Victims Killed': 'sum', 'Victims Injured': 'sum'})
victims_by_year.plot(kind='line', figsize=(10,6))
plt.title('Victims Killed and Injured Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Victims')
plt.show()

# 5. Investigate Missing Data (Suspect Information)
missing_suspects = merged_df[merged_df['Suspect Name'].isnull()]
missing_suspects_by_year = missing_suspects.groupby('Year').size()
missing_suspects_by_year.plot(kind='line', figsize=(10,6), color='orange')
plt.title('Missing Suspect Data Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Incidents with Missing Suspect Data')
plt.show()

# 6. Geospatial and Temporal Analysis (State vs Year)
state_year_incidents = merged_df.groupby(['State', 'Year']).size().unstack(fill_value=0)
plt.figure(figsize=(12,8))
sns.heatmap(state_year_incidents, cmap='coolwarm', cbar_kws={'label': 'Number of Incidents'})
plt.title('Mass Shootings by State and Year')
plt.xlabel('Year')
plt.ylabel('State')
plt.show()

# 7. Correlation Analysis (Time vs Victims)
correlation = merged_df[['Year', 'Victims Killed', 'Victims Injured']].corr()
print('Correlation Matrix:\n', correlation)

# 8. Time Series Forecasting (Using ARIMA for prediction)
# Forecasting mass shootings by year using ARIMA
model = ARIMA(yearly_incidents, order=(1, 1, 1))
model_fit = model.fit()
forecast = model_fit.forecast(steps=5)  # Forecasting next 5 years

print(f"Forecasted incidents for the next 5 years: {forecast}")

# Plotting forecast
plt.figure(figsize=(10,6))
plt.plot(yearly_incidents.index, yearly_incidents, label='Historical Data')
plt.plot(range(yearly_incidents.index[-1]+1, yearly_incidents.index[-1]+6), forecast, label='Forecast', color='red')
plt.title('Mass Shootings Forecast (Next 5 Years)')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.legend()
plt.show()

# Save all plots as images
plt.savefig('incident_frequency_over_time.png')
plt.savefig('seasonality_heatmap.png')
plt.savefig('day_of_week_analysis.png')
plt.savefig('victims_over_time.png')
plt.savefig('missing_suspects_over_time.png')
plt.savefig('state_year_analysis.png')
plt.savefig('time_series_forecast.png')

# Optionally, save the correlation matrix as a CSV
correlation.to_csv('correlation_matrix.csv')

print("Analysis Complete!")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Load the dataset
file_path = "weekly_aggregated_data.csv"  # Update with the correct file path if needed
weekly_agg = pd.read_csv(file_path)

# Convert 'Week Start' to datetime and extract the year
weekly_agg['Week Start'] = pd.to_datetime(weekly_agg['Week Start'])
weekly_agg['Year'] = weekly_agg['Week Start'].dt.year

# Aggregate yearly incidents for forecasting
yearly_data = weekly_agg.groupby('Year').sum(numeric_only=True)

# Define a function to forecast trends for different categories
def forecast_trend(column_name, train_until=2024, forecast_years=6):
    """Trains an Exponential Smoothing model and forecasts the given column."""
    # Filter data to train only on years up to the specified year
    train_data = yearly_data[column_name][yearly_data.index <= train_until]
    
    # Fit Exponential Smoothing model
    model = ExponentialSmoothing(train_data, trend="add", seasonal=None, damped_trend=True)
    fitted_model = model.fit()
    
    # Forecast the future years
    future_years = list(range(train_until + 1, train_until + 1 + forecast_years))
    forecast_values = fitted_model.forecast(steps=forecast_years)
    
    # Plot actual vs forecasted data
    plt.figure(figsize=(12, 6))
    plt.plot(train_data.index, train_data, label=f'Actual {column_name} (Until {train_until})', marker='o')
    plt.plot(future_years, forecast_values, label=f'Forecasted {column_name} ({train_until+1}-{train_until+forecast_years})', linestyle='dashed', marker='o', color='red')
    plt.xlabel('Year')
    plt.ylabel(column_name)
    plt.title(f'Forecast of {column_name} for {train_until+1}-{train_until+forecast_years}')
    plt.legend()
    plt.grid()
    plt.show()
    
    # Return the forecasted values as a DataFrame
    return pd.DataFrame({'Year': future_years, f'Forecasted {column_name}': forecast_values})

# Forecast for multiple categories
forecast_victims_killed = forecast_trend('Victims Killed')
forecast_victims_injured = forecast_trend('Victims Injured')
forecast_suspects_killed = forecast_trend('Suspects Killed')
forecast_suspects_injured = forecast_trend('Suspects Injured')
forecast_suspects_arrested = forecast_trend('Suspects Arrested')

# Combine all forecasts into one DataFrame
forecast_combined = pd.concat(
    [forecast_victims_killed.set_index('Year'),
     forecast_victims_injured.set_index('Year'),
     forecast_suspects_killed.set_index('Year'),
     forecast_suspects_injured.set_index('Year'),
     forecast_suspects_arrested.set_index('Year')],
    axis=1
).reset_index()

# Save the forecast data as a CSV file
forecast_combined.to_csv("forecasted_mass_shooting_trends_2025_2030.csv", index=False)

# Print the forecasted data
print("\nForecasted Mass Shooting Trends for 2025-2030:")
print(forecast_combined)
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.dynamic_factor import DynamicFactor
from statsmodels.tsa.api import VARMAX
import torch
import torch.nn as nn
from sklearn.preprocessing import MinMaxScaler

# Load the dataset
file_path = "weekly_aggregated_data.csv"
weekly_agg = pd.read_csv(file_path)

# Convert 'Week Start' to datetime and extract the year
weekly_agg['Week Start'] = pd.to_datetime(weekly_agg['Week Start'])
weekly_agg['Year'] = weekly_agg['Week Start'].dt.year

# Aggregate yearly incidents for multivariate forecasting
yearly_data = weekly_agg.groupby('Year')[['Victims Killed', 'Victims Injured', 'Suspects Killed', 'Suspects Injured', 'Suspects Arrested']].sum()

# Train using data until 2024
train_data = yearly_data.loc[yearly_data.index <= 2024].copy()

### 1. Dynamic Factor Model (DFM) ###
dfm_model = DynamicFactor(train_data, k_factors=1, factor_order=1)
dfm_fitted = dfm_model.fit()

# Forecast for 2025-2030
dfm_forecast = dfm_fitted.forecast(steps=6)
dfm_forecast.index = range(2025, 2031)

### 2. Vector Autoregressive Moving Average (VARMA) ###
varma_model = VARMAX(train_data, order=(1, 1))
varma_fitted = varma_model.fit()

# Forecast for 2025-2030
varma_forecast = varma_fitted.forecast(steps=6)
varma_forecast.index = range(2025, 2031)

### 3. LSTM-based Forecasting ###
# Normalize data
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(yearly_data)

# Convert to PyTorch tensors
X_train = torch.tensor(scaled_data[:-6], dtype=torch.float32)
y_train = torch.tensor(scaled_data[1:-5], dtype=torch.float32)

# Define LSTM model
class LSTMModel(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim, num_layers=2):
        super(LSTMModel, self).__init__()
        self.lstm = nn.LSTM(input_dim, hidden_dim, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_dim, output_dim)
    
    def forward(self, x):
        lstm_out, _ = self.lstm(x)
        return self.fc(lstm_out[:, -1, :])

# Train LSTM Model
input_dim = X_train.shape[1]
hidden_dim = 64
output_dim = input_dim
num_layers = 2
lstm_model = LSTMModel(input_dim, hidden_dim, output_dim, num_layers)

# Training loop
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(lstm_model.parameters(), lr=0.01)
X_train = X_train.unsqueeze(1)

for epoch in range(200):
    optimizer.zero_grad()
    output = lstm_model(X_train)
    loss = criterion(output, y_train)
    loss.backward()
    optimizer.step()

# Predict future values
X_future = torch.tensor(scaled_data[-6:], dtype=torch.float32).unsqueeze(1)
lstm_forecast_scaled = lstm_model(X_future).detach().numpy()

# Convert LSTM predictions back to original scale
lstm_forecast = scaler.inverse_transform(lstm_forecast_scaled)
lstm_forecast = pd.DataFrame(lstm_forecast, index=range(2025, 2031), columns=yearly_data.columns)

# Plot forecasts for all variables
variables = ['Victims Killed', 'Victims Injured', 'Suspects Killed', 'Suspects Injured', 'Suspects Arrested']

for var in variables:
    plt.figure(figsize=(12, 6))
    
    # Plot actual data till 2024
    plt.plot(train_data.index, train_data[var], label=f"Actual Data ({var})", marker='o')

    # Plot forecasts from models
    if var in dfm_forecast.columns:
        plt.plot(dfm_forecast.index, dfm_forecast[var], label="DFM Forecast", linestyle='dashed', marker='o')
    if var in varma_forecast.columns:
        plt.plot(varma_forecast.index, varma_forecast[var], label="VARMA Forecast", linestyle='dashed', marker='o')
    if var in lstm_forecast.columns:
        plt.plot(lstm_forecast.index, lstm_forecast[var], label="LSTM Forecast", linestyle='dashed', marker='o')

    # Configure plot
    plt.xlabel('Year')
    plt.ylabel(var)
    plt.title(f'Comparison of Forecasting Models for {var} (2025-2030)')
    plt.legend()
    plt.grid()
    plt.show()
```
